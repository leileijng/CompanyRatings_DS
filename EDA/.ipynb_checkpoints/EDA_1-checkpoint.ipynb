## 3. Exploring Ratings and Happiness with Other Features By Industry

### 3.1. Correlation between variables and rating by industry

```python
industries = ['Commodities', 'Financials', 'Govt_Services', 'Healthcare_NGO',
       'Leisure', 'Tech', 'Telecomm', 'Real_Estate', 'Retail', 'FNB']
variables = ['avg_salary', 'ceo_approval', 'Work/Life Balance', 'Compensation/Benefits', 'Job Security/Advancement', 
               'Management', 'Culture']

industryCorrDict = {}
for var in variables:
    industryCorrDict[var] = pd.DataFrame()
    
for industry in industries:
    df_temp = df[df['industry'] == industry]
    
    for var in variables:
        corr = df_temp['rating'].corr(df_temp[var])
        industryCorrDict[var][industry] = [corr]
```

```python
fig, axes = plt.subplots(4,2, figsize=(50,65))

sb.set_context("paper", 
               rc={"font.size":22,"axes.titlesize":30,"axes.labelsize":22, "xtick.labelsize":22, "ytick.labelsize": 22})

for i, var in enumerate(variables):
    row_index = int(i / 2)
    col_index = int(i - (row_index * 2) )
    chart = sb.barplot(data=industryCorrDict[var].reindex(industryCorrDict[var].mean().sort_values().index, axis=1), ax=axes[row_index, col_index], 
               orient='v')
    axes[row_index, col_index].set_title("Correlation of rating with " + var)
    chart.set_xticklabels(chart.get_xticklabels(), rotation=25)
    
```

### Discoveries:
1. While the overall correlation between **average salary** and **company rating** is quite low (`0.1399`), we can see that industry specific correlations tell a different story. Companies in the **Tecnology** and **Real Estate** industries have a moderate correlation of more than `0.3`. 

2. Surprisingly, this correlation is actually **negative** in the **Governmental Services** industry!
    - This suggests that different industries may value **salary** differently. 
    - This can also be seen in the gap for the correlation between **compensation/benefits** with **company rating**.
    
3. There is also a noticeable gap in the correlation between **CEO Approval** and **company rating** across industries, with the **Food and Beverages** industry at roughly `0.82` and the **Real Estate** industry at around `0.57`. 
    - One intuitive explanation could be the **level of autonomy** that employees have in their work. Industries with higher levels of autonomy (**Real Estate** and **Technology**) may not feel as much presence from top management compared to more **hierarchical** industries like **Governmental Services** and **Food and Beverages**. As such having a good management could be more important to these industries. 
    
4. As for the other factors, there is **no obvious differences** between industries.

### 3.2. Correlation between variables and happiness by industry

```python
happiness = ['ceo_approval', 'avg_salary', 'Work/Life Balance', 'Compensation/Benefits', 
             'Job Security/Advancement', 'Management', 'Culture', 'employees', 'revenue']
for var in happiness:
    industryCorrDict[var] = pd.DataFrame()
    
for industry in industries:
    df_temp = df2[df2['industry'] == industry]
    
    for var in happiness:
        corr = df_temp['Work Happiness Score'].corr(df_temp[var])
        industryCorrDict[var][industry] = [corr]
```

```python
fig, axes = plt.subplots(5,2, figsize=(50,65))

sb.set_context("paper", 
               rc={"font.size":22,"axes.titlesize":30,"axes.labelsize":22, "xtick.labelsize":22, "ytick.labelsize": 22})

for i, var in enumerate(happiness):
    row_index = int(i / 2)
    col_index = int(i - (row_index * 2) )
    chart = sb.barplot(data=industryCorrDict[var].reindex(industryCorrDict[var].mean().sort_values().index, axis=1), ax=axes[row_index, col_index], 
               orient='v')
    axes[row_index, col_index].set_title("Correlation of Work Happiness with " + var)
    chart.set_xticklabels(chart.get_xticklabels(), rotation=25)

    
    
```

### Discoveries:
1. We see a similar story here with **Work Happiness**, in that there are big differences in correlation with **average salary** across different industries. 
    - Correlation was negative for the **Food and Beverages**, **Governmental Services**, **Retail** and **Leisure** industries.
    - Again, **Real Estate** and **Technology** companies had the highest correlation between **average salary** and **work happiness**, above `0.3`.
    
2. This could also be seen in the correlation between **Compensations/Benefits** and **Work Happiness**, with **Technology** at the top with a correlation of `0.7` compared to **Food and Beverages** at `0.27`. 

3. The differences between industries are less pronounced in the remaining factors, albeit the differences are bigger than when we looked at correlation with **company rating**. 

4. This suggests that different industries value the various factors differently. 

5. It is also worth pointing out that companies from **Governmental Services** had the highest correlation for the remaining factors, although there are no obvious explanations.

### Discoveries (After PCA OLS Model):
1. Even though **work happiness** is mostly uncorrelated with **company size** (**employees** & **revenue**), there are noticeable differences across the various industry.
2. Therefore, **company size** when used together with **industry** could provide useful information in predicting **work happiness**.
